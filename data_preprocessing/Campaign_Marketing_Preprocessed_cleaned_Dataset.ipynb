# Cleaning and Preprocessing of Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import os
from sklearn import preprocessing as pp
from matplotlib import pyplot as plt
```

```python
print(os.listdir())
```

```python
df = pd.read_csv("Bank_marketing_cleaned_Data.csv")
df
```

```python
#Detecting NA,N/A and Null values
```

```python
df.isnull().sum()
```

```python
y_array =df[['y']]
y_array
```

```python
#Replacing 'yes' with 1 and 'no' with 0 in columb y
```

```python
y_array = y_array['y'].replace({'no': 0, 'yes': 1})
y_array
```

```python
df['y'] = y_array
df
```

```python
#Replacing 'single' with 0 , 'married' with 1 and 'divorced' with 2 in columb marital
```

```python
marital_array =df[['marital']]
marital_array = marital_array['marital'].replace({'single': 0,'married' : 1, 'divorced': 2})
df['marital'] = marital_array
df
```

```python
# Similarly replacing with binary values as above for columbs 'education','housing','loan','contact','poutcome'
```

```python
y_array =df[['education']]
y_array = y_array['education'].replace({'primary': 0,'secondary' : 1, 'tertiary': 2})
df['education'] = y_array
df
```

```python
y_array =df[['housing']]
y_array = y_array['housing'].replace({'yes': 1,'no' : 0})
df['housing'] = y_array
df
```

```python
y_array =df[['loan']]
y_array = y_array['loan'].replace({'yes': 1,'no' : 0})
df['loan'] = y_array
df
```

```python
y_array =df[['contact']]
y_array = y_array['contact'].replace({'cellular': 1,'telephone' : 0})
df['contact'] = y_array
df
```

```python
y_array =df[['poutcome']]
y_array = y_array['poutcome'].replace({'success': 1,'failure' : 0,'other' : 2})
df['poutcome'] = y_array
df
```

```python
#Using 3 standard deviation to remove the outliers
```

```python
upper_limit= balance_array.balance.mean() + 3*balance_array.balance.std()
upper_limit
```

```python
lower_limit = balance_array.balance.mean() - 3*balance_array.balance.std()
lower_limit
```

```python
balance_array[(balance_array.balance>upper_limit) | (balance_array.balance<lower_limit)]
```

```python
#Romoving these 53 outliers to balance the data
```

```python
df=df[(df.balance<upper_limit) & (df.balance>lower_limit)]
```
